Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular data visualization library in Python that provides a wide range of tools for creating static, animated, and interactive plots. It is widely used in various fields, including data analysis, scientific research, and software development.

Matplotlib allows users to create high-quality graphs, figures, and visualizations in a simple and flexible manner. It provides a comprehensive set of functions and classes for creating a variety of plots, charts, histograms, scatter plots, and much more.

The Pyplot module is a sub-library of Matplotlib that provides a simplified interface for creating plots. It is commonly imported using the plt alias. Here are five types of plots that can be created using the Pyplot module:

1)Line Plot: A basic plot that displays data points connected by straight lines, useful for visualizing trends or patterns over time.


2) Bar Plot: A plot that represents data with rectangular bars, commonly used to compare categorical data or display frequencies.


3) Scatter Plot: A plot that displays individual data points as markers, allowing the examination of relationships between two variables.


4) Histogram: A graphical representation of the distribution of numerical data, dividing the data into bins and displaying the frequency of each bin.


5) Pie Chart: A circular statistical graphic that represents the proportions of different categories within a dataset, typically used to show percentages or proportions.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.


A scatter plot is a type of plot that displays individual data points as markers on a two-dimensional coordinate system. 

It is particularly useful for visualizing the relationship or correlation between two variables. Each data point in a scatter plot is represented by a marker, such as a dot or a circle, with its position determined by its corresponding x and y values.




```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
```

import numpy as np

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to create multiple plots in a single figure window, organizing them in rows and columns.



The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the subplot you want to create. The index starts from 1 and increases from left to right and top to bottom.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')

# Adjust the layout and display the plot
plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])

A bar plot, also known as a bar chart, is a type of plot that presents categorical data with rectangular bars. Each bar represents a category, and the length or height of the bar corresponds to the value or quantity associated with that category. Bar plots are commonly used to compare and visualize data across different categories or groups.

```python
import numpy as np
import matplotlib.pyplot as plt 
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel('Company')
plt.ylabel('Profit ')
plt.title('Company Profits')
plt.show()
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset's summary statistics. It displays the distribution of numerical data through quartiles, outliers, and other statistical measures. Box plots are commonly used to understand the spread, skewness, and potential outliers within a dataset.




Box plots are used for various purposes, including:

Summarizing data: Box plots provide a concise summary of the dataset's distribution, showing the range, central tendency (median), and variability (IQR).
Comparing groups: Box plots are effective in comparing multiple datasets or groups side by side, allowing quick comparisons of their medians, quartiles, and ranges.
Detecting outliers: The presence of outliers can be visually identified in box plots, helping to detect potential anomalies or data points that deviate significantly from the bulk of the data.
Visualizing distributions: Box plots provide a visual representation of the shape and symmetry of a dataset's distribution, such as skewness or the presence of multiple peaks.





```python
import numpy as np
import matplotlib.pyplot as plt
a =np.random.normal(10,100,20)
b =np.random.normal(20,90,20)
data = [a,b]
plt.figure(figsize=(8, 5))
plt.boxplot(data)
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()
```
